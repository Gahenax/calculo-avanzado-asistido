# 🌌 Mersenne Gap Observatory: Spectral Entropy Analysis

Welcome to the **Mersenne Gap Observatory**.
This notebook provides the tools to explore the cryptographic entropy and "Ghost Loci" density surrounding the known Mersenne Primes ($M_p = 2^p - 1$).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="darkgrid", palette="viridis")
plt.rcParams['figure.figsize'] = (12, 6)
```

## 1. Load the Telemetry Dataset
We load the structural data of the first 51 known Mersenne neighborhoods.

```python
try:
    df = pd.read_parquet('../data/mersenne_telemetry_v1.parquet')
except:
    df = pd.read_csv('../data/mersenne_telemetry_v1.csv')

display(df.head())
print(f"Loaded {len(df)} High-Precision Mersenne Environments.")
```

## 2. Ghost Loci Density vs Exponent ($p$)
Ghost Loci represent "near-prime" structural anomalies detected by the OEDA pipeline prior to a certified Lucas-Lehmer discovery.

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x='exponent_p', y='ghost_loci_density', data=df, s=100, color='crimson', edgecolor='black')
plt.title('Ghost Loci Density scaling with $p$ in $M_p = 2^p-1$')
plt.xlabel('Exponent $p$')
plt.ylabel('Detected Ghost Loci')
plt.xscale('log')
plt.yscale('log')
plt.show()
```
